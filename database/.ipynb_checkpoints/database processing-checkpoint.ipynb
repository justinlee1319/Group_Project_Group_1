### About Our Dataset
This dataset is mainly about top streamers on twitch from past year.  
This data consists of different things like number of viewers, number of active viewers, followers gained and many other relevant columns regarding a particular streamer. It has 11 different columns, 7 of them are integers, 2 of them are strings, and the rest of 2 are Booleans. So, this dataset has 1000 rows and 11 columns, it is suitable for our group project.  
Reference link: https://www.kaggle.com/datasets/aayushmishra1512/twitchdata

### Data collection

```python
# import libraries and modules we will use
import pandas as pd
import sqlite3
import csv 
```

```python
# use pandas to read the csv file on local disk.
csv_path = r'E:\Data Analytics\DAB111 python\group project\my zip\my zip\twitchdata-update.csv'
df = pd.read_csv(csv_path)
```

```python
# since this data looks quite clean and clear, there is no need to do any data cleaning process.
df.head()
```

### Database processing

```python
# create a database using sqlite3
con = sqlite3.connect('Twitch_data.db')
cursor = con.cursor()
```

```python
# create a table named "twitch_data", and import columns and values from our local disk stored csv file "twitchdata-update.csv" into table.
df.to_sql('twitch_data', con, index=False, if_exists='replace')
```

```python
# let us check whether the data we inserted is successful stored or not.
df_from_sql = pd.read_sql('SELECT * FROM twitch_data', con)
df_from_sql
```

```python
# save what we created and exsit databse.
con.commit()
con.close
```
